graficos de series de tiempo

```python
import pandas as pn
```

```python

data_range = pn.date_range(start='2023-01-01',end='2023-12-31',freq='D')
precios = np.cumsum(np.random.randn(len(data_range))) + 100
```

```python
data = pn.DataFrame({'Data':data_range,'Precios':precios})
plt.plot(data['Data'],data['precios'],label='Precio')
plt.show
```

# Ejercicio 2

```python
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.cos(x)

categorias = ['a','b','c','d']
valores = [5,7,8,3]

a  = np.random.rand(90)
b = np.random.rand(90)

datos = np.random.randn(10000)
```

```python
fig, axs = plt.subplots(2,2,figsize= (15,10))

axs[0,0].plot(x,y1,label = 'seno')
axs[0,0].plot(x,y2,label = 'coseno')
axs[0,0].set_title('Grafico de lineas')
axs[0,0].legend(loc='upper right')
axs[0,0].grid(True)

axs[0,1].bar(categorias,valores)
axs[0,1].set_title('Grafico de barras')

axs[1,0].scatter(a,b,color='r', alpha=.5)
axs[1,0].set_title('Grafico de dispersion')

axs[1,1].hist(datos,bins=30)
axs[1,1].set_title('Grafico de Histograma')

fig.suptitle('Ejemplos de diferentes tipos de graficos',fontsize = 20)
plt.show
```
