La reduccin de los niveles de cuantizacin convierte una imagene con una escala de grises de X niveles de intesidad a otra con una escala menor, reduciendo el volumen de datos

```python
def gray2ind(img, k):
    """
    Cuantiza una imagen en escala de grises a 'k' niveles distintos.

    Parmetros:
    - img (numpy.ndarray): Imagen en escala de grises de entrada.
    - k (int): Nmero de niveles de grises deseados despus de la cuantizacin.

    Retorna:
    - numpy.ndarray: Imagen cuantizada.
    """

    # Calcula el intervalo de niveles de grises.
    # Por ejemplo, para una imagen de 8 bits (256 niveles) y k=4, el nivel ser 64.
    nivel = 256 // k

    # Cuantiza la imagen:
    # - Primero, divide cada pxel de la imagen por 'nivel'.
    # - Luego, multiplica el resultado por 'nivel' para obtener el valor cuantizado.
    # - Finalmente, convierte los valores a tipo uint8 (imagen de 8 bits).
    im1 = (img // nivel * nivel).astype(np.uint8)

    return im1
```

Qu es un histograma?

Un histograma de una imagen representa la distribucin de intensidades de pxeles en esa imagen. Si una imagen tiene la mayora de sus pxeles concentrados en un rango estrecho de intensidades, puede parecer demasiado oscura o demasiado clara.

```python

def mostrar_histograma(img, title="Histograma"):
    plt.hist(img.ravel(),              , [0,256])
    plt.title(title)
    plt.xlabel('Intensidad')
    plt.ylabel('Nmero de Pxeles')
    plt.show()
```

La ecualizacin del histograma es especialmente til en imgenes con bajo contraste, ya que puede hacer que los detalles que anteriormente estaban ocultos se vuelvan visibles.

## Thresholding (Umbralizacin)



El "thresholding" o umbralizacin es una tcnica de procesamiento de imgenes que se utiliza para binarizar una imagen. Esta tcnica convierte una imagen en una compuesta nicamente por dos valores de pxeles, generalmente representados como blanco y negro.

La idea bsica detrs de la umbralizacin es establecer un valor de umbral. Luego:

- Todos los pxeles en la imagen con un valor **por encima** de ese umbral se configuran a un valor (por lo general blanco).
- Todos los pxeles con un valor **por debajo** de ese umbral se configuran a otro valor (generalmente negro).

La umbralizacin se describe mediante la siguiente frmula:

\[ f(x, y) =
\begin{cases}
\text{maxValue} & \text{si } I(x, y) > \text{umbral} \\
\text{minValue} & \text{si } I(x, y) \leq \text{umbral}
\end{cases}
\]

Donde:
- \( f(x, y) \) es el valor del pxel en la imagen de salida en la posicin (x, y).
- \( I(x, y) \) es el valor del pxel en la imagen de entrada en la posicin (x, y).
- "maxValue" y "minValue" son, respectivamente, los valores mximos y mnimos que puede tener un pxel en la imagen. En imgenes binarias, "maxValue" suele ser 255 (blanco) y "minValue" es 0 (negro).

### Tcnicas de Umbralizacin

1. **Umbralizacin Simple**: Se elige un valor de umbral fijo.
2. **Umbralizacin Adaptativa**: El valor de umbral se determina de manera adaptativa para diferentes regiones de la imagen. Esto es til cuando hay variaciones de iluminacin en la imagen.
3. **Umbralizacin de Otsu**: Esta tcnica elige automticamente el valor de umbral que maximiza la varianza entre dos clases de pxeles (fondo y primer plano).

### Aplicaciones de la Umbralizacin

La umbralizacin es til en tareas como:
- **Segmentacin de Imgenes**: Donde se desea separar objetos de inters del fondo.
- **Preprocesamiento**: Para mejorar la deteccin de bordes o contornos.
- **Extraccin de Caractersticas**: De la imagen.

> **Nota**: OpenCV, una popular biblioteca de procesamiento de imgenes, proporciona funciones como `cv2.threshold` y `cv2.adaptiveThreshold` para realizar operaciones de umbralizacin en imgenes.

## Operadores morfologicos

Las operaciones morfolgicas son tcnicas fundamentales en el procesamiento de imgenes binarias (y a veces tambin en imgenes en escala de grises). Se basan en la modificacin de formas en una imagen mediante la aplicacin de un elemento estructurante. Los principales tipos de operaciones morfolgicas son:

- Erosin
- Dilatacin
- Apertura
- Cierre

Erosin

La erosin "erosiona" o reduce las regiones brillantes (blancas) de la imagen. Es til para eliminar pequeas manchas o estructuras no deseadas.

```python
kernel = np.ones((3,3), np.uint8)
eroded_image =                   (gaussianAdaptiveThresh, kernel, iterations=1)
mostrar_imagen(eroded_image, "Image")
```

Dilatacin

La dilatacin "expande" o aumenta las regiones brillantes (blancas) de la imagen. Es til para conectar objetos o estructuras cercanas.

```python
kernel = np.ones((3,3), np.uint8)
dilated_image =                   (gaussianAdaptiveThresh, kernel, iterations=1)
mostrar_imagen(dilated_image, "Image")
```

Apertura

La apertura es una erosin seguida de una dilatacin. Es til para eliminar el ruido o pequeas manchas de una imagen.

```python
kernel = np.ones((3,3), np.uint8)
opened_image = cv2.morphologyEx(gaussianAdaptiveThresh, cv2.MORPH_OPEN, kernel)
mostrar_imagen(opened_image, "Image")
```

Cierre

El cierre es una dilatacin seguida de una erosin. Es til para cerrar pequeos agujeros o huecos en objetos.

```python
kernel = np.ones((3,3), np.uint8)
closed_image = cv2.morphologyEx(gaussianAdaptiveThresh, cv2.MORPH_CLOSE, kernel)
mostrar_imagen(closed_image,"Image")
```

Momentos de Hu


Los momentos de Hu son un conjunto de 7 momentos invariantes (o nmeros) que son invariantes a las transformaciones de traslacin, escala y rotacin. Estos momentos son ampliamente utilizados en el reconocimiento de patrones y anlisis de formas en imgenes, ya que proporcionan una descripcin numrica de la forma de un objeto, que sigue siendo la misma independientemente de su orientacin, tamao o posicin.

Dado que son invariantes a estas transformaciones, se utilizan a menudo para reconocer y clasificar formas y objetos en imgenes, incluso si estn rotados, escalados o trasladados.

https://es.wikipedia.org/wiki/Momentos_de_imagen


```python

import numpy as np

# Suponiendo que tienes una imagen binarizada 'binary_image'
contours, _ = cv2.findContours(otsuThresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

hu_moments_list = []

for contour in contours:
    # Calcula los momentos normales del contorno
    moments = cv2.moments(contour)

    # Calcula los momentos de Hu a partir de los momentos normales
    hu_moments = cv2.HuMoments(moments)

    # Opcional: si quieres transformar los momentos para mejorar la comparacin
    log_hu_moments = [-np.sign(m)*np.log10(np.abs(m)) for m in hu_moments]

    hu_moments_list.append(log_hu_moments)

# Ahora, 'hu_moments_list' contendr una lista de los 7 momentos de Hu para cada contorno
print(hu_moments_list)
```
